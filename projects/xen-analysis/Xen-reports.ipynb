## Query to be used

```python
query = """ SELECT message_ID,
                   first_date,
                   subject,
                   message_body
            FROM messages
            WHERE subject like '%[PATCH%'
        """
data = execute_query(cursor, query)
           
```

## Casting the data to Pandas

It seems that there are around 140k mails dealing with patchsets, or at least with the key word "PATCH" in their subjects

```python
data_list = list(data)
len(data)
```

```python
import pandas

pandas_ds = pandas.DataFrame(data_list, columns=["message_id", "first_date", "subject", "message_body"])
```

```python
# Remove not needed information in the subject
pandas_ds["subject"].replace('.*\[Xen-devel\].*\[PATCH', '[PATCH', inplace=True, regex=True)
```

```python
# Let's see how the subject looks like now
pandas_ds["subject"]
```

```python
pandas_ds["original_subject"] = pandas_ds["subject"]
```

```python
pandas_ds["original_subject"]
```

```python
# Let's remove the patch part and count patchsets with same subject
pandas_ds["subject"].replace('.*\[PATCH.*\]', '', inplace=True, regex=True)
```

```python
# This should be a close number to the total number of patchsets, counting as a patchset any of the subparts of a patch series
pandas_ds["subject"].nunique()
```

```python
# This action group by the same subject the list of emails and provides the first and last date of that subject appearance in the mailing list
# together with the number of times that subject takes place in the mailing list
import numpy as np
pandas_ds.groupby("subject").agg({"first_date": [max, min], "subject": len})
```
