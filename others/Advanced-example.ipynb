```python
import pandas

pandas_ds = pandas.DataFrame(data_list, columns=["added_lines", "removed_lines", "date", "author", "repository", "company"])

pandas_ds.columns.values.tolist()
```

```python
pandas_ds
```

```python
# unique number of companies
pandas_ds.company.nunique()
```

```python
# how to access a specific column through its name (same result as above)
pandas_ds["company"].nunique()
```

```python
# unique number of companies, and all added lines
import numpy as np
pandas_ds.groupby("company").agg({"added_lines" : np.sum,
                                "removed_lines" : np.sum,
                                "repository" : len})
```

```python
# group by company
companies = pandas_ds.groupby("company")

# group by repository
repositories = pandas_ds.groupby("repository")

# group by developer
developers = pandas_ds.groupby("author")

com_repo = pandas_ds.groupby(["company", "repository"])
```

```python
companies
```

```python
#added lines per company
companies.aggregate({"removed_lines" : sum})
```

```python
# aggregation of removed_lines by company and repository
com_repo.aggregate({"author" : pandas.Series.nunique,
                    "added_lines" : sum})
```

```python
# aggregation of all metrics
import numpy as np
com_repo.aggregate(np.sum)
```

```python
import numpy as np
#mean added lines per company
# from http://pandas.pydata.org/pandas-docs/stable/groupby.html
# Aggregating functions are ones that reduce the dimension of the returned objects, 
# for example: mean, sum, size, count, std, var, sem, describe, first, last, nth, min, max
companies.aggregate({"removed_lines" : np.size})
```
