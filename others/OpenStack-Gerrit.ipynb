The following query provides information about the gerrit_issue, summary of the changeset, submitter, opening date, number of patchsets, gerrit tracker, closing date, time to close and current status.

```python
query = """ select i.issue                       as gerrit_issue, 
                   i.summary                     as summary, 
                   i.submitted_by                as changeset_submitter, 
                   t2.opening_date               as opening_date, 
                   count(distinct(ch.old_value)) as num_patchsets, 
                   t.url                         as gerrit_tracker, 
                   t1.closed_date                as closed_date, 
                   timestampdiff(SECOND, t2.opening_date, t1.closed_date) as time2close, 
                   i.status                      as current_status 
            from issues i, 
                 trackers t, 
                 changes ch, 
                 (select i.id as issue_id, 
                         ch.changed_on as closed_date 
                  from issues i 
                    left join changes ch on ch.issue_id = i.id and 
                                            field='status' and 
                                            (new_value='ABANDONED' or new_value='MERGED') ) t1, 
                 (select ch.issue_id, 
                         ch.changed_on as opening_date 
                  from changes ch 
                  where ch.field='status' and 
                        ch.new_value='UPLOADED' and 
                        ch.old_value=1) t2 
            where i.tracker_id=t.id and 
                  ch.issue_id = i.id and 
                  i.id=t1.issue_id  and 
                  i.id=t2.issue_id 
            group by i.issue """

```

```python
data = execute_query(cursor, query)
```

## Importing data from MySQL to Pandas

```python
data_list = list(data)
```

```python
import pandas

pandas_ds = pandas.DataFrame(data_list, columns=["gerrit_issue", "summary", "changeset_submitter", "opening_date", "num_patchsets", "gerrit_tracker", "closed_date", "time2close", "current_status"])

pandas_ds.columns.values.tolist()
```

## Grouping in Pandas

```python
# This uses the groupby functiona and apply an operator to the rest of the columns.

# Let's group by gerrit_tracker
```

```python
pandas_ds.groupby("gerrit_tracker").agg({"changeset_submitter" : pandas.Series.nunique})
```

```python
# And we can keep adding other columns to the agg function. We need to detail the function to be applied in each case.
import numpy as np
pandas_ds.groupby("gerrit_tracker").agg({"changeset_submitter" : pandas.Series.nunique,
                                          "num_patchsets" : np.mean })
```
